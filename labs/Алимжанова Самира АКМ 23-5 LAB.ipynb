### Задание 3: Циклы и условные операторы (Максимум: 1.5 балла)

**Ваша задача:** Выполните сортировку данных по категориям, используя цикл `for` и конструкцию `if/elif/else`.

*   **Вариант 1:** Дан список `temperatures = [-4, 25, 12, 18, 0, 33, -1, 15]`. Создайте три новых списка: `freezing` (температуры ниже 0), `cool` (от 0 до 19 вкл.) и `warm` (выше 19).

*   **Вариант 2:** Дан список `words = ['analytics', 'data', 'science', 'ml', 'python', 'ai', 'model']`. Создайте два новых списка: `short_words` (слова длиной до 3 символов вкл.) и `long_words` (остальные).

*   **Вариант 3:** Дан список `scores = [59, 91, 82, 75, 65, 40, 99, 78, 89]`. Создайте три списка оценок: `C` (60-79), `B` (80-89) и `A` (90+). Баллы ниже 60 игнорируйте.

*   **Вариант 4:** Дан список `numbers = [10, 21, 35, 40, 55, 64, 77, 80, 93]`. Создайте два списка: `divisible_by_5` и `divisible_by_10` (числа, которые делятся на 5 и 10 без остатка соответственно).

> **Подсказки:**
> * Сначала создайте пустые списки для результатов.
> * В цикле `for` проверяйте каждый элемент и добавляйте его в нужный список методом `.append()`.

```python
# ВАРИАНТ-1
temperatures = [-4, 25, 12, 18, 0, 33, -1, 15]
freezing=[]
cool=[]
warm=[]
for tem in temperatures:
    if tem < 0:
        freezing.append(tem)
    elif tem <= 19:
        cool.append(tem)
    else:
        warm.append(tem)
print("Freezing:", freezing)
print("Cool:", cool)
print("Warm:", warm)
```

*   **Вариант 4:** Напишите функцию `get_grade(score)`. Если `score` >= 90, **вернуть** 'A'; 75-89 — 'B'; 60-74 — 'C'; иначе — 'F'. Функция должна **вернуть** строковую оценку.

### Задание 6: Итоговая задача (Максимум: 2.5 балла)

**Ваша задача:** Напишите одну функцию, которая решает комплексную задачу из вашего варианта.

*   **Вариант 1:** Напишите функцию `get_top_performers(employees, performance_threshold)`. На вход: список словарей `employees` (`'name'`, `'department'`, `'performance_score'`) и порог. Функция должна вернуть список **имен** сотрудников из отдела `'Sales'`, чей `performance_score` выше или равен порогу.

*   **Вариант 2:** Напишите функцию `find_available_books(books, min_copies)`. На вход: список словарей `books` (`'title'`, `'genre'`, `'copies_available'`) и мин. кол-во. Функция должна вернуть список **названий** книг в жанре `'Fantasy'`, у которых `copies_available` >= `min_copies`.

*   **Вариант 3:** Напишите функцию `get_active_projects(projects, status)`. На вход: список словарей `projects` (`'project_id'`, `'team_lead'`, `'status'`) и `status`. Функция должна вернуть список **ID проектов**, которые имеют заданный `status`.

*   **Вариант 4:** Напишите функцию `analyze_student_grades(students, min_attendance)`. На вход: список словарей `students` (`'name'`, `'attendance_percentage'`, `'average_grade'`) и мин. посещаемость. Функция должна вернуть список **имен** студентов с `attendance_percentage` > `min_attendance` и `average_grade` > 4.0.

> **Подсказки:**
> * План: создайте пустой список, в цикле `for` перебирайте словари, внутри цикла с помощью `if` проверяйте все условия (используйте `and`), и если они верны, добавляйте нужное значение в список.

```python
# Вариант -1
def get_top_performers(employees, performance_threshold):
    top_performers = []  # создаём пустой список для результатов

    for employee in employees:  # перебираем каждого сотрудника
        # проверка условий: отдел — 'Sales', и балл >= порога
        if employee['department'] == 'Sales' and employee['performance_score'] >= performance_threshold:
            top_performers.append(employee['name'])  # добавляем имя в список

    return top_performers  # возвращаем список имён
```
